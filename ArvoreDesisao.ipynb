```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

import pandas as pd

# Local dataset no git
#url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/datasets/iris.data" # Modelo Professor Dr. Thiago Moura
#url = "https://raw.githubusercontent.com/otanerifpb/dataset/main/iris.data" # Modelo Git Renato
#url = "https://raw.githubusercontent.com/otanerifpb/dataset/main/diabetes1.data" # Teste dataset 1 - diabetes
url = "https://raw.githubusercontent.com/otanerifpb/dataset/main/metropt.data" # Teste dataset 2 - metropt

# Carregar base de dados
# DataFrame
dataset = pd.read_csv(url, header=None)

columns = len(dataset.columns)

y = dataset[0] # extrai a primeira coluna, que é o label
X = dataset.loc[:,1:columns-1]
x1 = dataset
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste
```

```python
# Treinamendo da Árvore de Decisão
#model = tree.DecisionTreeClassifier(criterion="entropy",max_depth=2)
model = tree.DecisionTreeClassifier(criterion="entropy")
model = model.fit(X_train, y_train)
```

```python
# Predição e Resultados

result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```
